```python
array = df["views"].to_numpy()
array
```

```python
average_views = np.mean(array)
median_views = np.median(array)
maximum_views = np.max(array)
minimum_viewss = np.min(array)
standard_deviation = np.std(array)
print(average_views)
print(median_views)
print(maximum_views)
print(minimum_viewss)
print(standard_deviation)
```
